```python
import os

# Define the path to your data folder
data_dir = r'C:\Users\C1\Desktop\NLP Project\train - Copy'

# List all files in the data directory
files = os.listdir(data_dir)

# Load all text files into a list
data = []
for file in files:
    file_path = os.path.join(data_dir, file)
    with open(file_path, 'r', encoding='utf-8') as f:
        data.append(f.read())

# Now, `data` contains all the loaded text data from the files
print(f"Loaded {len(data)} files.")
print(data[:100])
```

```python
import re

data = str(data)
def clean_text(text):
  cleaned = re.sub(r'[^\u0900-\u097F\s]', '', text)

  cleaned = re.sub(r'\s+', ' ', cleaned).strip()
  return cleaned

cleaned_corpus = clean_text(data)

print(cleaned_corpus[:10000])
```

```python
tokens = cleaned_corpus.split()
print(tokens[:100])
```

```python
from collections import defaultdict

def generate_ngrams(tokens, n):
    ngrams = defaultdict(list)

    for i in range(len(tokens) - n):
        key = tuple(tokens[i:i+n-1])
        value = tokens[i+n-1]
        ngrams[key].append(value)

    return ngrams

trigrams = generate_ngrams(tokens, 3)

for key, value in list(trigrams.items())[:5]:
    print(f'{key}: {value}')
```

```python
import random

def predict_next_word(input_text, ngrams):
    input_tokens = input_text.split()

    if len(input_tokens) < 2:
        return "Input too short"

    key = tuple(input_tokens[-2:])
    if key in ngrams:
        return random.choice(ngrams[key])
    else:
        return "No prediction available"

# Example
input_text =  "यह शहर बोबर"
predicted_word = predict_next_word(input_text, trigrams)
print(f'Predicted next word: {predicted_word}')
```

```python
import random

def predict_next_word(input_tokens, ngrams):
    # Get the last two words as key for trigram model
    key = tuple(input_tokens[-2:])

    # Check if the key exists in the trigram model
    if key in ngrams:
        # Randomly choose one of the possible next words
        return random.choice(ngrams[key])
    else:
        return None

def generate_sentence(input_text, ngrams, num_words=5):
    input_tokens = input_text.split()  # Tokenize the input

    # Ensure the input has at least two words for a trigram model
    if len(input_tokens) < 2:
        return "Input too short to predict"

    # Generate words to extend the sentence
    for _ in range(num_words):
        next_word = predict_next_word(input_tokens, ngrams)

        # Stop if no next word is found
        if not next_word:
            break

        # Add the predicted word to the current sentence
        input_tokens.append(next_word)

    # Join the tokens to form a complete sentence
    return ' '.join(input_tokens)

# Example usage
input_text = "यह शहर बोबर"
generated_sentence = generate_sentence(input_text, trigrams, num_words=5)
print(f'Generated sentence: {generated_sentence}')
```
